# Scaling the data for machine learning

```python
# Remove columns that don't require scaling
df_sub = df.drop(columns=['DATE', 'MONTH'])
```

```python
# Check for missing values
df_sub.isnull().sum()
```

```python
# Eliminate records that have NaN values
df_sub.dropna(inplace=True)
```

```python
df_sub.shape
```

```python
# Create scaler object using StandardScaler from sklearn.preprocessing
scaler = StandardScaler()
```

```python
# Create new df with scaled data
columns = df_sub.columns
df_scaled = pd.DataFrame(scaler.fit_transform(df_sub), columns=columns)
df_scaled.head()
```

```python
# Compare to the original data
df_sub.head()
```

```python
# Export the scaled data frame
df_scaled.to_csv(os.path.join(path, 'Dataset-weather-prediction-dataset-processed_scaled.csv'))
```
